```python
def list_match(gold_list, prop_list, row_id, rank_type):
    match_cnt = 0
    total_cnt = 0
    for elem in prop_list:
        for gold in gold_list:
            #print(gold, elem)
            if fuzz.partial_ratio(gold, elem) > 90:
                match_cnt = match_cnt + 1
                #print(gold)
                #print('********', elem)
                break
        total_cnt = total_cnt + 1
    print('Row id: ', row_id, 'Rank type: ' , rank_type, ' Matched: ', match_cnt, ' Total count: ', total_cnt)
    if total_cnt != 3:
        print('Row id: ', row_id, 'Rank type: ' , rank_type, ' has less than 3 elements returned.')
    #print(, total_cnt)
    #print("__________________________")
    return match_cnt
```

We now study those data points that are misclassified by all the models. We map them to the sentence sequence ids, which we then use to associate with the different sentence-level aspect scores.

```python
error_all_model_data = pd.read_csv('Rcodes/SentAllModelFail_4thJuly.csv')
sent_map_data = pd.read_csv('Rcodes/FinalMappedAnnotatedINFOsents_17thSept.csv')
```

```python
print(error_all_model_data.columns)
print(sent_map_data.columns)
```

```python
error_gold_map = list()
error_sent_seq_id_map = list()
error_authid_list = list()
```

```python
for row_one in range(error_all_model_data.shape[0]):
    error_authid = sent_data.iloc[error_all_model_data.iloc[row_one, 0], 0]
    for row_two in range(sent_map_data.shape[0]):
        if sent_map_data.iloc[row_two, 0] == error_authid:
            if fuzz.partial_ratio(error_all_model_data.iloc[row_one, 1], sent_map_data.iloc[row_two, 1]) > 90:
                print(error_all_model_data.iloc[row_one, 1], '\n ************ \n', sent_map_data.iloc[row_two, 1], '_________________________________________--')
                error_gold_map.append(error_all_model_data.iloc[row_one, 1])
                error_sent_seq_id_map.append(sent_map_data.iloc[row_two, 2])
                error_authid_list.append(error_authid)
                break
```

```python
error_map_data = pd.DataFrame({'X.AUTHID': error_authid_list, 'sent_seq_id': error_sent_seq_id_map, 'gold_sent': error_gold_map})
error_map_data.to_csv('ErrorAllModel_SentIdMap4thJune.csv', index=False)
error_map_data.head()
```
